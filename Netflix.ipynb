### Plot the count of different types of shows availble on netflix

```python
#Plot the count of different types of shows availble on netflix
count_data = df['type'].value_counts()
count_data.plot(kind='bar',color=['Red','Black'])
plt.title('Count of Different Shows on Netflix')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

### Top 10 category with the Most Content on Netflix

```python
# Extracting unique categories from the 'listed_in' column
all_categories = ', '.join(df['listed_in']).split(', ')
```

```python
# Create a pandas Series to count occurrences of each category
category_counts = pd.Series(all_categories).value_counts()
```

```python
# Get the top 10 categories with the most content
top_categories = top_categories.sort_values(ascending=True)
```

```python
# Get the top 10 categories with the most content
top_categories = category_counts.head(10)
```

```python
print(top_categories)
```

```python
# Create a horizontal bar plot using matplotlib
plt.barh(top_categories.index, top_categories.values)

# Customize the plot
plt.xlabel('Number of Contents')
plt.ylabel('Category')
plt.title('Top 10 Categories with the Most Content on Netflix')

# Display the plot
plt.show()
```
